```python
# 1. Two Sum
class Solution:
    def twoSum(self, nums: list[int], target: int) -> list[int]:
        for i in range(len(nums)):
            for j in range(i+1, len(nums)):
                if nums[i] + nums[j] == target:
                    return [i, j]
```

```python
# 2. Add Two Numbers
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def addTwoNumbers(self, l1: ListNode, l2: ListNode) -> ListNode:
        
        val = (l1.val + l2.val) % 10
        bonus = (l1.val + l2.val) // 10
        res = ListNode(val, None)
        true_res = res
        
        l1 = l1.next
        l2 = l2.next
        
        while l1 != None or l2 != None:
            two_sum = []
            if l1 != None:
                two_sum.append(l1.val)
            if l2 != None:
                two_sum.append(l2.val)
            val = (sum(two_sum) + bonus) % 10
            bonus = (sum(two_sum) + bonus) // 10
            res.next = ListNode(val, None)
            res = res.next
            if l1 != None:
                l1 = l1.next
            if l2 != None:
                l2 = l2.next

        if bonus == 1:
            res.next = ListNode(1, None)
               
        return true_res
```

```python
# 4. Median of Two Sorted Arrays
class Solution:
    def findMedianSortedArrays(self, nums1: list[int], nums2: list[int]) -> float:
        nums = sorted([*nums1, *nums2])
        if len(nums) % 2 == 0:
            val1 = nums[len(nums) // 2]
            val2 = nums[len(nums) // 2 - 1]
            return (val1 + val2) / 2
        else:
            return nums[len(nums) // 2]
```

```python
# 7. Reverse Integer
class Solution:
    def reverse(self, x: int) -> int:
        x_str = str(x)
        if x_str[0] == '-':
            x_str = '-' + x_str[:0:-1]
        else:
            x_str = x_str[::-1]
        x = int(x_str)
        if x < -(2**31) or x > (2**31)-1:
            return 0
        else:
            return x
```

```python
# 8. String to Integer (atoi)
class Solution:
    def myAtoi(self, s: str) -> int:
        result = ''
        stage = 1
        for char in s:
            if stage == 1:
                if char == ' ':
                    continue
                else:
                    stage = 2
            if stage == 2:
                stage = 3
                if char in ['+', '-']:
                    result += char
                    continue
            if stage == 3:
                if char.isdigit():
                    result += char
                    continue
                else:
                    break
        if result in ['', '+', '-']:
            return 0
        else:
            positive = True
            if result[0] in ['+', '-']:
                sign = result[0]
                result = result[1:]
                if sign == '-':
                    positive = False
            result = int(result)
            if positive == False:
                result = result*(-1)
            if result < -2**31:
                result = -2**31
            elif result > (2**31)-1:
                result = (2**31)-1
            else:
                pass
            return result
```

```python
# 24. Swap Nodes in Pairs
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def swapPairs(self, head: ListNode = None) -> ListNode or None:
        
        if head == None:
            return None
        elif head.next == None:
            return head
        else:
            if head.next.next != None:
                prev = head
            answ = head.next
            head.next = head.next.next
            answ.next = head
            head = head.next
        
        while head != None:
            temp = head.next
            if head.next == None:
                break
            elif head.next.next == None:
                prev.next = head.next
                head.next = head.next.next
                temp.next = head
                break
            elif head.next.next.next == None:
                prev.next = head.next
                head.next = head.next.next
                temp.next = head
                break
            else:
                prev.next = temp
                head.next = head.next.next
                temp.next = head
                prev = head
                head = head.next
        
        return answ
```

```python
# 43. Multiply Strings
class Solution:
    def multiply(self, num1: str, num2: str) -> str:
        integers = []
        for string in [num1, num2]:
            integers.append(0)
            length = len(string)
            for char in string:
                integers[-1] += int(char)*(10**(length-1))
                length -= 1
        result = integers[0]*integers[1]
        return str(result)
```

```python
# 167. Two Sum II - Input Array Is Sorted
class Solution:
    def twoSum(self, numbers: list[int], target: int) -> list[int]:
        for i in range(len(numbers)):
            for j in range(i+1, len(numbers)):
                two_sum = numbers[i] + numbers[j]
                if two_sum == target:
                    return [i+1, j+1]
                elif two_sum > target:
                    break
```
